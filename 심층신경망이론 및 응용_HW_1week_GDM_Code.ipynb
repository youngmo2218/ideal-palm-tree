```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sympy import Derivative, symbols
%matplotlib inline
```

```python
# input data
X=[0,1,1,2]
Y=[0,1,2,1]
```

```python
#구하고자 하는 가중치 바이어스를 문자로 변경
W2=symbols('W2')
W1=symbols('W1')
W0=symbols('W0')
```

```python
#loss function 계산함수
def lossfuntion(x,y):
    e=0
    for i in range(len(x)):
        a=x[i]
        b=y[i]
        c=W2*a**2+W1*a+W0
        d=(b-c)**2
        e=e+d
        print(e)
    return e
```

```python
#Loss Funtion 함수 도출
MSE=lossfuntion(X,Y)
#Lossfunction 을 각 변수별 편미분 수행
MSE_W2 = Derivative(MSE,W2).doit()
MSE_W1 = Derivative(MSE,W1).doit()
MSE_W0 = Derivative(MSE,W0).doit()
```

```python
#초기 조건 및 학습률, 학습 횟수 지정
Learning_Rate = 0.001
Learning_Count = 1000000
a=2
b=3
c=2
a_new=2
b_new=3
c_new=2
gradient__=[]
```

```python
#학습 Elary Stop 기능, W0,1,2 값 DataFrame 변경
for i in range(1):
    gradient_=[]
    for k in range(Learning_Count):
        value={W0:a_new,W1:b_new,W2:c_new}
        a_new = a - Learning_Rate*MSE_W0.subs(value)
        b_new = b - Learning_Rate*MSE_W1.subs(value)
        c_new = c - Learning_Rate*MSE_W2.subs(value)
        if abs(a_new-a) > 0.00001 or  abs(b_new-b) > 0.00001 or abs(c_new-c) > 0.00001:
            a=a_new
            print('a_new',a_new)
        
        #if abs(b_new-b) > 0.00001:
            b=b_new
            print('b_new',b_new)
        #if abs(c_new-c) > 0.00001:
            c=c_new
            print('c_new',c_new)
        else:
            break
        gradient = [a_new,b_new,c_new,k+1]
        gradient__.append(gradient)
        print('횟수=',k+1)

        #a=a_new
        #b=b_new
        #c=c_new

```
